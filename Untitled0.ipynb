```python
# Задача №39.
# Даны два массива чисел. Требуется вывести те элементы первого массива
# (в том порядке, в каком они идут в первом массиве), которых нет во втором массиве.
# Пользователь вводит число N - количество элементов в первом массиве,
# затем N чисел - элементы массива. Затем число M - количество элементов во втором массиве.
# Затем элементы второго массива
# Ввод:                       Вывод
# 7                         3 3 2 12
# 3 1 3 4 2 4 12
# 6
# 4 15 43 1 15 1            (каждое число вводится с новой строки)

n = int(input())      #  Вводим кол-во элементов 1го массива
list1 = list()        #  Создаем пустой список list1
for i in range(n):    #  заполняем его с помощью цикла
    x = int(input())  #  при каждой итерации будем вводить наше значение х с помощью input
    list1.append(x)   #  добавляем этот х в список list1 c помощью функции append

     # Все то же самое для второго массива

m = int(input())      #  Вводим кол-во элементов 2го массива
list2 = list()        #  Создаем пустой список list2
for i in range(m):    #  заполняем его с помощью цикла
    x = int(input())  #  при каждой итерации будем вводить наше значение х с помощью input
    list2.append(x)   #  добавляем этот х в список list1 c помощью функции append

count = 0                         #  создаем счетчик где будем хранить повторения, если нет то ставим 0,
for i in range(n):                #  цикл до n для 1го массива
    for j in range(m):            #  цикл до m для 2го массива
        if list1[i] == list2[j]:  #  каждый элемент 1го массива будем сравнивать с каждым элементом 2го массива
            count += 1            #  если нашли совпадающие элементы, то count увеличиваем на ед.
    if count == 0:                #  если не найдено совпадений и count равна 0, то
        print(list1[i])           #  выводим элементы 1го массива
    count = 0                     #  и  count = 0


```

```python
# Задача №41.
# Дан массив, состоящий из целых чисел. Напишите программу, которая в данном массиве определит
# количество элементов, у которых два соседних и, при этом, оба соседних элемента меньше данного.
# Сначала вводится число N — количество элементов в массиве
# Далее записаны N чисел — элементы массива. Массив состоит из целых чисел.
# Ввод:                       Ввод
# 5                            5
# 1 2 3 4 5                 1 5 1 5 1
# Вывод                       Вывод
# 0                            2


n = int(input())         #  вводим кол-во элементов списка
list1 = list()           #  создаем список с помощью функции list
for i in range(n):       #  с помощью цикла for заполняем ее элементами
    x = int(input())      #  каждый раз вводим наше число
    list1.append(x)       #   и добавлять его в список

count = 0                  #  создаем переменную счетчик, чтобы мы могли посчитать сколько
                           # элементов подходят под наши условия
for i in range(1, n - 1):  #  создаем цикл с диапазоном под наши условия (1, n-1)
    if list1[i - 1] < list1[i] > list1[i + 1]:  # если ай минус первый элемент (i-1) меньше чем итый элемент [1]
                          # и одновременно он больше чем i + 1й элемент [i+1]
        count += 1         #  если это так, то мы счетчик увеличиваем не 1 ед.
print(count)               #
```

```python

# Задача №43.
# Дан список чисел. Посчитайте, сколько в нем пар элементов, равных друг другу.
# два элемента, равные друг другу образуют одну пару, которую необходимо посчитать.
# Вводится список чисел. Все числа списка находятся на разных строках.
# Ввод:                    Вывод:
# 1 2 3 2 3                  2

list1 = [1, 2, 3, 2, 3]                   #  Создадим список
count = 0                                 #  создаем счетчик в котором будем записывать количество пар

for i in range(len(list1)):                # создаем for в котором проходим по всему списку
    for j in range(i+1, len(list1)):       # в этом же цикле создаем еще один цикл, где первый элемент будет
                                           # сравниваться с другими элементами до конца list1.
        if i != j and list1[i] == list1[j]:    # если выбранные элементы не одинаковы (i != j) и одновременно с этим
                                           # другие элементы наоборот одинаковы лист итое == листу джитое, то
            count += 1                     #  счетчик увеличиваем на 1.
print(count)                               # результат 2

```

```python
# Задача №45.
# Два различных натуральных числа n и m называются дружественными, если сумма делителей числа n
# (включая 1, но исключая само n) равна числу m и наоборот.
# Например, 220 и 284 – дружественные числа.
# По данному числу k выведите все пары дружественных чисел, каждое из которых не превосходит k.
# Программа получает на вход одно натуральное число k, не превосходящее 105 .
# Программа должна вывести все пары дружественных чисел, каждое из которых не превосходит k.
# Пары необходимо выводить по одной в строке, разделяя пробелами. Каждая пара должна быть
# выведена только один раз (перестановка чисел новую пару не дает).
# Ввод:                Вывод:
# 300                  220 284
#


k = int(input())                 #  Создаеи переменную для ввода кол-ва элементов
list1 = list()                   #  создаем список состоящий из кортежа. 1й элемент это число, а второе это
                                 #  сумма его делителя

for i in  range(k):               #  создаем цикл где будем перебирать все наши значения до К
    summa = 0                     #  создаем переменную суума и обяжем что она равняется 0 и будет обновляться при каждой итерации
    for j in range(1,i//2 + 1):   #  создаем второй цикл где будем проверять наши делители, делятся они на 2 иои нет
                                  #  и будем идти от 1 до i//2 + 1.
        if i % j == 0:            #  если наше i ай делится без остатка
            summa += j              #  то нашу сумму увеличиваем на наш делитель j
    list1.append(tuple([i,summa]))   # и в список list1 добавлять значения и создаем кортеж с помощью функции tuple
                                     # ставим i и сумму ее делителя.
    # если сейчас выведем через print(list1) получим связку число и его делитель.

for i in range(len(list1)):          # через цикл будем искать i элементы которые у нас совпадают
    for j in range(i, len(list1)):    #  через цикл будем искать j элементы которые у нас совпадают
        if i != j and list1[i][0] == list1[j][1] and list1[i][1] == list1[j][0]:
            # проверяем все значения i и j крест на крест
            print(*list1[i])         # выводим с распаковкой кортежа (звездочка)

```
